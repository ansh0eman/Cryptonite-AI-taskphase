## Handling missing values

```python
# check the number of missing values in each column
data.isnull().sum()
```

```python
# drop the cabin column 
data = data.drop(columns= 'Cabin', axis =1)
```

We want to fill in missing age data instead of just dropping the missing age data rows. One way to do this is by filling in the mean age of all the passengers (imputation). However we can be smarter about this and check the average age by passenger class.

```python
plt.figure(figsize= (10,7))
sns.boxplot(x= 'Pclass', y = 'Age', data = data, palette='winter')
```

We can see the wealthier passengers in the higher classes tend to be older, which makes sense. We'll use these average age values to impute based on Pclass for Age.

```python
def impute_age(df):

    for pclass in df['Pclass'].unique():

        median_age = df[df['Pclass']== pclass]['Age'].median()
        df.loc[(df['Pclass']== pclass) & (df['Age'].isnull()), 'Age'] = median_age

    return df

# OR

# replacing the missing values in "Age" column with median value passengerclass wise 

# data['Age'].fillna(data.groupby('Pclass')['Age'].transform('median'), inplace=True)
```

```python
impute_age(data)
```

```python
## replace with mean fare of Pclass 3
data.Fare.fillna(data[data['Pclass']== 3]['Fare'].mean(), inplace= True)
```

```python
# replacing the missing values in "Embarked" column with mode value

data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

## splitting into train and test datasets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## model training 
**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegressionCV, LogisticRegression

model_1 = LogisticRegression(random_state= 2)

model_2 = LogisticRegressionCV(cv = 5, random_state= 2)
```

```python
# training the Logistic Regression model with training data

model_1.fit(X_train, Y_train)

model_2.fit(X_train, Y_train)
```

## Model Evaluation
**Accuracy Score**

```python
# accuracy on training data

X_train_prediction_1 = model_1.predict(X_train)

X_train_prediction_2 = model_2.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score

training_data_accuracy_1 = accuracy_score(Y_train, X_train_prediction_1)
print('Accuracy score of training data on Logistic Regression model : ', training_data_accuracy_1)

training_data_accuracy_2 = accuracy_score(Y_train, X_train_prediction_2)
print('Accuracy score of training data on Logistic Regression CV model (cv=5): ', training_data_accuracy_2)
```

```python
# accuracy on test data

X_test_prediction_1 = model_1.predict(X_test)

X_test_prediction_2 = model_2.predict(X_test)
```

```python
test_data_accuracy_1 = accuracy_score(Y_test, X_test_prediction_1)
print('Accuracy score of test data Logistic Regression model: ', test_data_accuracy_1)

test_data_accuracy_2 = accuracy_score(Y_test, X_test_prediction_2)
print('Accuracy score of test data Logistic Regression CV model: ', test_data_accuracy_2)
```

## F1 Score

```python
from sklearn.metrics import f1_score

f1_train = f1_score(Y_train, X_train_prediction_1)
print(f'F1 score for Training data : {f1_train}')

f1_test = f1_score(Y_test, X_test_prediction_1)
print(f'F1 score for Test data : {f1_test}')
```
