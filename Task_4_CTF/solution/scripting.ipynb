```python
from tensorflow.keras.models import load_model

# Load the provided model
model = load_model('../my_model.keras')
```

```python
# View the model's architecture
model.summary()

#Identify the layers where the activations are used (focus on hidden layers).
```

Decode The architecture

List out the activation functions in the hidden layers.

Translate each ReLU to 1 and Sigmoid to 0.

```python
for layer in model.layers:
    print(layer.get_config())
```

```python
activations = []

for layer in model.layers:
    # Check if the layer is an activation layer
    if 'activation' in layer.name or 're_lu' in layer.name:
        activation_function = layer.get_config().get('activation', 'relu')  # Defaults to 'relu' for re_lu layers
        if activation_function == 'relu':
            activations.append('1')
        elif activation_function == 'sigmoid':
            activations.append('0')

# Combine the activations to form the binary flag
binary_flag = ''.join(activations)
print("Binary Flag:", binary_flag)
```

```python
def binary_to_ascii(binary_str):
    ascii_text = ''.join(chr(int(binary_str[i:i+8], 2)) for i in range(0, len(binary_str), 8))
    return ascii_text

# Decode the flag
flag = binary_to_ascii(binary_flag)
print("Hidden Flag:", flag)
```

## Explanation

Layer Inspection: The script now correctly checks for activation layers where ReLU (re_lu) and Sigmoid are used as separate layers. This is why it works correctly for your model setup.

Binary Conversion: The script converts the activation sequence (ReLU to 1, Sigmoid to 0) into a binary string, which is then decoded into the ASCII flag.
